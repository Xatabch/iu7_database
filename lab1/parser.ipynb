```python
import json
from pprint import pprint
import random
```

```python
json_data=open("winemag-data.json").read()

data = json.loads(json_data)
pprint(data)
```

```python
newData = data[:100]
```

```python
json1 = []

for i in range(len(newData)):
    json1.append({})
    json1[i]['id'] = i + 1
    json1[i]['title'] = newData[i]['title']
    json1[i]['variety'] = newData[i]['variety']
    json1[i]['price'] = random.randint(10,20) if newData[i]['price'] == None else newData[i]['price']
    json1[i]['country_id'] = tp(newData[i]['country'])
    json1[i]['description_id'] = i + 1
    
```

```python
def tp(strk):
    if(strk == "Argentina"):
        n = 7
    elif(strk == "Australia"):
        n = 9
    elif(strk == "Austria"):
        n = 10
    elif(strk == "Chile"):
        n = 8
    elif(strk == "France"):
        n = 5
    elif(strk == "Germany"):
        n = 6
    elif(strk == "Italy"):
        n = 1
    elif(strk == "Portugal"):
        n = 2
    elif(strk == "Spain"):
        n = 4
    elif(strk == "US"):
        n = 3
    return n  
```

```python
file = open("first_essense.txt","w")
for i in range(len(json1)):
    file.write(str(json1[i]['id'])+ "|" + str(json1[i]['title']) + "|" + str(json1[i]['variety']) + "|" + str(json1[i]['price']) + "|" + str(json1[i]['country_id']) + "|" + str(json1[i]['description_id']) + "\n")
    
file.close()
```

```python
json2 = []
copy = []
j = 0

for i in range(len(newData)):
    if newData[i]['taster_name'] not in copy and newData[i]['taster_name'] != None:
        json2.append({})
        json2[j]['id'] = j + 1
        json2[j]['taster_name'] = newData[i]['taster_name']
        json2[j]['taster_twitter_handle'] = newData[i]['taster_twitter_handle']
        copy.append(newData[i]['taster_name'])
        j+=1

file = open("second_essense.txt","w")
for i in range(len(json2)):
    file.write(str(json2[i]['id']) + "|" + str(json2[i]['taster_name']) + "|" + str(json2[i]['taster_twitter_handle']) + "\n")
    
file.close()
```
